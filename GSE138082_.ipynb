```r

install.packages("data.table")
install.packages("R.utils")
if (!requireNamespace("BiocManager", quietly = TRUE)) install.packages("BiocManager")
BiocManager::install("DESeq2", ask = FALSE)
BiocManager::install("apeglm", ask = FALSE)

library(data.table)
library(R.utils)
library(DESeq2)
library(apeglm)


gunzip("GSE138082_Subfields_Counts.txt.gz", remove = FALSE, overwrite = TRUE)
counts_data <- fread("GSE138082_Subfields_Counts.txt")
head(counts_data)
gene_names <- counts_data[[1]]
counts_matrix <- as.matrix(counts_data[, -1, with = FALSE])
rownames(counts_matrix) <- gene_names

dim(counts_matrix)
head(counts_matrix[, 1:5])
sample_names <- colnames(counts_matrix)
condition <- ifelse(grepl("CTL", sample_names, ignore.case = TRUE), "Control", "SCZ")
condition <- factor(condition)
colData <- data.frame(row.names = sample_names,
                      condition = condition)

keep_genes <- rowSums(counts_matrix) >= 10
filtered_counts <- counts_matrix[keep_genes, ]

cat("Before filtering:", nrow(counts_matrix), "\n")
cat("After filtering :", nrow(filtered_counts), "\n")
dds <- DESeqDataSetFromMatrix(countData = filtered_counts,
                              colData = colData,
                              design = ~ condition)

dds <- DESeq(dds)
res <- results(dds, contrast = c("condition", "SCZ", "Control"))
res <- lfcShrink(dds, coef="condition_SCZ_vs_Control", res=res)
sig_genes <- res[which(res$padj < 0.05), ]
cat("Number of significant genes (SCZ vs Control):", nrow(sig_genes), "\n")
head(sig_genes[order(sig_genes$padj), ])
res_df <- as.data.frame(res)
res_df$gene <- rownames(res_df)
res_df$significant <- "Not Significant"
res_df$significant[res_df$padj < 0.05 & abs(res_df$log2FoldChange) > 1] <- "Significant"
library(ggplot2)
ggplot(res_df, aes(x = log2FoldChange, y = -log10(pvalue), color = significant)) +
  geom_point(alpha = 0.6) +
  scale_color_manual(values = c("grey", "red")) +
  theme_minimal() +
  labs(title = "Volcano Plot: SCZ vs Control",
       x = "Log2 Fold Change",
       y = "-Log10(p-value)")

install.packages("pheatmap")
library(pheatmap)
sig_genes <- res[which(res$padj < 0.05), ]
sig_genes <- sig_genes[order(sig_genes$padj), ]
top_genes <- rownames(sig_genes)[1:50]
vsd <- vst(dds, blind = FALSE)
norm_counts <- assay(vsd)

heatmap_matrix <- norm_counts[top_genes, ]

sample_conditions <- as.character(colData$condition)
names(sample_conditions) <- rownames(colData)
annotation_col <- data.frame(Condition = sample_conditions)
rownames(annotation_col) <- colnames(heatmap_matrix)

pheatmap(heatmap_matrix,
         scale = "row",
         annotation_col = annotation_col,
         show_rownames = TRUE,
         show_colnames = TRUE,
         clustering_distance_rows = "euclidean",
         clustering_distance_cols = "euclidean",
         clustering_method = "complete",
         fontsize = 8,
         main = "Top 50 DE Genes (SCZ vs Control)")

if (!requireNamespace("BiocManager", quietly = TRUE))
  install.packages("BiocManager")
BiocManager::install("topGO", ask = FALSE, update = FALSE)
BiocManager::install("org.Hs.eg.db", ask = FALSE, update = FALSE)

library(topGO)
library(org.Hs.eg.db)
geneList <- as.numeric(res$padj < 0.05)
names(geneList) <- rownames(res)
geneList[is.na(geneList)] <- 0

geneSymbols <- names(geneList)
entrez_ids <- mapIds(org.Hs.eg.db, keys = geneSymbols,
                     column = "ENTREZID", keytype = "SYMBOL", multiVals = "first")


valid_genes <- !is.na(entrez_ids)
geneList <- geneList[valid_genes]
names(geneList) <- entrez_ids[valid_genes]
GOdata <- new("topGOdata",
              ontology = "BP",
              allGenes = geneList,
              geneSelectionFun = function(x) x == 1,
              annot = annFUN.org,
              mapping = "org.Hs.eg.db",
              ID = "entrez")

resultFisher <- runTest(GOdata, algorithm = "classic", statistic = "fisher")
go_results <- GenTable(GOdata, classicFisher = resultFisher,
                       orderBy = "classicFisher", topNodes = 10)

print(go_results)
```
